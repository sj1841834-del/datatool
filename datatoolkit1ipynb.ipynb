### 21. Why is NumPy's array slicing faster than Python's list slicing?

NumPy's array slicing is faster than Python's list slicing due to:

- *Contiguous memory allocation*: NumPy arrays are stored in contiguous blocks of memory, allowing efficient access and manipulation.
- *Low-level optimizations*: NumPy is implemented in C, which enables faster execution of operations compared to Python’s native list handling.
- *Vectorized operations*: Slicing in NumPy doesn’t create copies unless explicitly requested; it returns views, reducing memory overhead and speeding up computation.
- *Homogeneous data types*: NumPy arrays have fixed data types, allowing optimized CPU instructions, unlike Python lists which can hold mixed types.

```python
#4. Calculate correlation matrix using Seaborn and display heatmap
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1,2,3,4,5],
    'B': [5,3,6,2,7],
    'C': [2,4,1,8,3]
})
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```
